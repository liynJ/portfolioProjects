```python
import pandas as pd
import os
```

```python
pwd = os.getcwd()
```

```python
dataset = pd.read_csv( pwd + "/orders_from_20210831_to_20211002_20211019_1109.csv")
dataset
```

```python
dataset_modified = dataset.copy()
```

```python
dataset_modified.dtypes
```

```python
dataset_modified['PO Number'] = dataset_modified['PO Number'].str.lstrip('="')
dataset_modified['PO Number'] = dataset_modified['PO Number'].str.rstrip('"')
dataset_modified
```

```python
list = (dataset_modified.columns)
list
```

```python
columns_to_drop = ['Order Quantity', 'Account Group','Currency', 'Order Subtotal',
       'Order Shipping & Handling', 'Order Promotion', 'Order Tax','Approver', 'Account User',
       'Account User Email', 'Invoice Status', 'Total Amount',
       'Invoice Due Amount', 'Invoice Issue Date', 'Invoice Due Date','Payment Instrument Type', 'Payment Identifier',
       'Amazon-Internal Product Category','UNSPSC','Segment', 'Family', 'Class', 'Commodity', 'Brand Code', 'Brand',
       'Manufacturer', 'National Stock Number', 'Item model number',
       'Part number', 'Product Condition', 'Company Compliance', 'Listed PPU',
       'Purchase PPU', 'Item Subtotal',
       'Item Shipping & Handling', 'Item Promotion', 'Item Tax','PO Line Item Id', 'Tax Exemption Applied',
       'Tax Exemption Type', 'Tax Exemption Opt Out', 'Discount Program',
       'Pricing Discount applied ($ off)', 'Pricing Discount applied (% off)',
       'GL Code', 'Department', 'Cost Center', 'Project Code', 'Location',
       'Custom Field 1', 'Seller Name', 'Seller Credentials', 'Seller City',
       'Seller State', 'Seller ZipCode']
columns_to_drop
```

```python
dataset_modified = dataset_modified.drop(columns = columns_to_drop)
dataset_modified
```

```python
drop_rows = dataset_modified[(dataset_modified['Order Status'] == 'Cancelled') & (dataset_modified['Payment Reference ID'].isna())].index
drop_rows
```

```python
dataset_modified.drop(drop_rows, inplace=True)
```

```python
dataset_modified['amazon_join'] = dataset_modified['PO Number'] + dataset_modified['ASIN']
```

```python
dataset_modified['Order Date'] = pd.to_datetime(dataset_modified['Order Date'], format='%m%d%Y')
```

```python
start_date = '2021-08-31'
end_date = '2021-10-01'
mask = (dataset_modified['Order Date'] > start_date) & (dataset_modified['Order Date'] < end_date)
dataset_modified = dataset_modified.loc[mask]
```
